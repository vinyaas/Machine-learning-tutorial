# Data File Types DS

## Text Files
### CSV (Comma-Separated Values)
- **Usage**: Storing tabular data.
- **Scenario**: Importing data into pandas for data analysis.
- **Example**: Customer data, sales data.

### TSV (Tab-Separated Values)
- **Usage**: Similar to CSV but uses tabs.
- **Scenario**: Data where commas are part of the dataset.
- **Example**: Bioinformatics data.

### TXT (Plain Text File)
- **Usage**: Storing unstructured text.
- **Scenario**: Natural Language Processing (NLP).
- **Example**: Corpus of text documents.

### JSON (JavaScript Object Notation)
- **Usage**: Data exchange between servers and web apps.
- **Scenario**: API responses.
- **Example**: User profile information from a web service.

## Excel Files
### XLS/XLSX
- **Usage**: Spreadsheets with multiple sheets and formulas.
- **Scenario**: Business analytics and reporting.
- **Example**: Financial statements, budget sheets.

## Binary Files
### Parquet
- **Usage**: Columnar storage optimized for big data.
- **Scenario**: Big data processing with Apache Spark.
- **Example**: Large-scale transaction logs.

### Avro
- **Usage**: Serialization format, row-oriented.
- **Scenario**: Data exchange and storage in Hadoop.
- **Example**: Log data.

### Feather
- **Usage**: Fast read/write operations for pandas and R.
- **Scenario**: Quick data transfer between systems.
- **Example**: Intermediate data analysis steps.

## Database Formats
### SQL (Structured Query Language)
- **Usage**: Relational database storage.
- **Scenario**: Complex queries and data relationships.
- **Example**: Customer orders database.

### HDF5 (Hierarchical Data Format)
- **Usage**: Large and complex datasets.
- **Scenario**: Scientific computing and research.
- **Example**: Genomic data, large image datasets.

## Image Files
### JPEG/JPG
- **Usage**: Compressed images.
- **Scenario**: Computer vision projects.
- **Example**: Image classification datasets.

### PNG
- **Usage**: Lossless compressed images.
- **Scenario**: Data visualization and plotting.
- **Example**: Plot images from matplotlib.

### TIFF
- **Usage**: High-quality images.
- **Scenario**: Medical imaging.
- **Example**: MRI scan images.

### GIF
- **Usage**: Animated graphics.
- **Scenario**: Web data visualization.
- **Example**: Animated data plots.

## Audio/Video Files
### WAV/MP3
- **Usage**: Audio data.
- **Scenario**: Speech recognition.
- **Example**: Voice command datasets.

### MP4/MOV
- **Usage**: Video data.
- **Scenario**: Video analysis.
- **Example**: Surveillance footage.

## Specialized Formats
### MAT (MATLAB)
- **Usage**: Data exchange with MATLAB.
- **Scenario**: MATLAB-based projects.
- **Example**: Simulation data.

### NPY (NumPy)
- **Usage**: Storing NumPy arrays.
- **Scenario**: Python-based data processing.
- **Example**: Intermediate results in a machine learning pipeline.

### ARFF (Attribute-Relation File Format)
- **Usage**: Weka machine learning tool.
- **Scenario**: Weka-based machine learning tasks.
- **Example**: Classification datasets.
