# K-Means Clustering Algorithm

## Introduction
K-means clustering is a popular unsupervised machine learning algorithm used to partition a dataset into `k` clusters, where each data point belongs to the cluster with the nearest mean (centroid). It is widely used for its simplicity and efficiency.

## How It Works
The k-means algorithm follows these steps:
1. **Initialization**: Choose the number of clusters `k` and randomly initialize the centroids.
2. **Assignment**: Assign each data point to the nearest centroid, forming `k` clusters.
3. **Update**: Calculate the new centroids as the mean of all data points in each cluster.
4. **Repeat**: Repeat the assignment and update steps until the centroids no longer change or a maximum number of iterations is reached.

## Steps to Use K-Means Clustering
1. **Import Libraries**:
    ```python
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn.cluster import KMeans
    ```

2. **Load Data**:
    ```python
    # Example data
    X = np.array([[1, 2], [1, 4], [1, 0], 
                  [4, 2], [4, 4], [4, 0]])
    ```

3. **Initialize and Fit K-Means**:
    ```python
    kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
    ```

4. **Predict Clusters**:
    ```python
    labels = kmeans.predict(X)
    ```

5. **Visualize Clusters**:
    ```python
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
    plt.show()
    ```

## Applications of K-Means Clustering
1. **Customer Segmentation**: Grouping customers based on purchasing behavior for targeted marketing.
2. **Image Compression**: Reducing the number of colors in an image by clustering similar colors.
3. **Document Clustering**: Organizing a large set of documents into topics.
4. **Anomaly Detection**: Identifying unusual data points in a dataset.

## Real-Life Case Scenarios
1. **Retail**: A retail company uses k-means clustering to segment customers into different groups based on their purchase history and preferences. This helps in creating personalized marketing strategies.
2. **Healthcare**: Hospitals use k-means clustering to group patients with similar symptoms or medical histories, aiding in better diagnosis and treatment plans.
3. **Finance**: Banks use k-means clustering to detect fraudulent transactions by identifying patterns that deviate from normal behavior.

## Conclusion
K-means clustering is a versatile and powerful algorithm for partitioning data into meaningful clusters. Its simplicity and efficiency make it a popular choice for various applications across different domains.



````python
import pandas as pd 
import numpy as np
````

````python
# Set a seed for reproducibility
np.random.seed(0)

# Create a DataFrame with 50 rows and three features: 'Age', 'Income' and 'Score'
data = {
    'Age': np.random.randint(20, 50, 50),
    'Income': np.random.randint(10000, 50000, 50),
    'Score': np.random.randint(1, 100, 50)
}
df = pd.DataFrame(data)
df
````

````python

from sklearn.cluster import KMeans

#specifying the number of cluster (k) u want to divide ur data into 
k = 3 

kmeans = KMeans(n_clusters=4 , random_state= 42)
kmeans.fit(df)
print(kmeans.labels_)
````

````python
import matplotlib.pyplot as plt

# Scatter plot of the data before clustering
plt.figure(figsize=(6, 6))
plt.scatter(df['Age'], df['Income'], c='blue')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Data before clustering')
plt.show()
````

````python
# Scatter plot of the data after clustering
plt.figure(figsize=(6, 6))
plt.scatter(df['Age'], df['Income'], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Data after clustering')
plt.show()
````

````python
df['cluster'] = kmeans.labels_

# Group by cluster and calculate mean
cluster_centers = df.groupby('cluster').mean()
cluster_centers
````

````python
from sklearn.preprocessing import MinMaxScaler

# Initialize a scaler
scaler = MinMaxScaler()

# Fit and transform the data
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Calculate mean values for each cluster
cluster_characteristics_scaled = df_scaled.groupby('cluster').mean()

# Plotting the cluster characteristics
cluster_characteristics_scaled.plot(kind='bar', figsize=(10,7))
plt.title('Mean Characteristics of Each Cluster (Normalized)')
plt.ylabel('Mean Value (Normalized)')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
````
